```
#boxen plot script

import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


def cat_plot(data, title, x_axis, x_label, y_axis,y_label, y_lim, rot=0, cols=None):
  plt.close()
  plt.clf()
  g = sns.catplot(x=x_axis, y=y_axis, kind='boxen', col=cols, data=data)
  g.set(title = title)
  g.set(ylim=(0, y_lim))
  g.set(ylabel = y_label)
  g.set(xlabel = x_label)
  plt.xticks(rotation=rot)
  return g.fig
```
